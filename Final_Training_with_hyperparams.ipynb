```python
a = a.assign(popularity='hit')
b = b.assign(popularity='popular')
c = c.assign(popularity='not_hit')
d = d.assign(popularity='flop')
```

```python
df=pd.concat([a,b,c,d])
```

```python
X = df.drop('popularity', axis=1)  # Features
y = df['popularity']  # Target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, stratify=y, random_state=42)

# Apply SMOTE to the training set only
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Initialize and train the Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train_resampled, y_train_resampled)

# Make predictions on the test set
y_pred = rf_classifier.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Generate and print the classification report
class_report = classification_report(y_test, y_pred)
print('Classification Report:\n', class_report)
```

```python
import pandas as pd
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Assuming df contains your features and target variable
X = df.drop('popularity', axis=1)  # Features
y = df['popularity']  # Target variable

# Initialize the Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Perform stratified k-fold cross-validation (let's use 5-fold as an example)
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Initialize lists to store cross-validation scores and predictions
cv_scores = []
cv_predictions = []

# Perform cross-validation without data leakage
for train_idx, test_idx in kfold.split(X, y):
    X_train_fold, X_test_fold = X.iloc[train_idx], X.iloc[test_idx]
    y_train_fold, y_test_fold = y.iloc[train_idx], y.iloc[test_idx]
    
    # Apply SMOTE to the training fold
    smote = SMOTE(random_state=42)
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train_fold, y_train_fold)
    
    # Train the Random Forest classifier on the resampled data
    rf_classifier.fit(X_train_resampled, y_train_resampled)
    
    # Make predictions on the test fold
    y_pred_fold = rf_classifier.predict(X_test_fold)
    
    # Calculate accuracy for this fold and append to the list of scores
    accuracy_fold = accuracy_score(y_test_fold, y_pred_fold)
    cv_scores.append(accuracy_fold)
    
    # Store predictions for later analysis if needed
    cv_predictions.extend(y_pred_fold)

# Calculate mean accuracy across folds
mean_accuracy = sum(cv_scores) / len(cv_scores)
print(f'Mean Cross-Validated Accuracy: {mean_accuracy:.2f}')


```

```python
from sklearn.preprocessing import LabelEncoder

# Convert string labels to numerical classes using LabelEncoder
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the encoded labels into training and testing sets
X_train, X_test, y_train_encoded, y_test_encoded = train_test_split(X, y_encoded, test_size=0.20, stratify=y_encoded, random_state=42)

# Apply SMOTE to the training set only
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train_encoded)
```

```python
import xgboost as xgb

# Initialize and train the XGBoost classifier
xgb_classifier = xgb.XGBClassifier(random_state=42)
xgb_classifier.fit(X_train_resampled, y_train_resampled)

# Make predictions on the test set
y_pred_xgb = xgb_classifier.predict(X_test)

# Calculate accuracy for XGBoost
accuracy_xgb = accuracy_score(y_test_encoded, y_pred_xgb)
print(f'XGBoost Accuracy: {accuracy_xgb:.2f}')

# Generate and print the classification report for XGBoost
class_report_xgb = classification_report(y_test_encoded, y_pred_xgb)
print('XGBoost Classification Report:\n', class_report_xgb)
```

```python
import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint, uniform

# Define the parameter distributions to sample from
param_dist = {
    'n_estimators': randint(100, 1000),  # Number of boosting rounds
    'max_depth': randint(3, 10),  # Maximum depth of the tree
    'learning_rate': uniform(0.01, 0.3)  # Step size shrinkage used to prevent overfitting
}

# Create an XGBoost classifier
xgb_classifier = xgb.XGBClassifier(random_state=42)

# Initialize RandomizedSearchCV with the classifier, parameter distributions, and number of iterations
random_search = RandomizedSearchCV(estimator=xgb_classifier, param_distributions=param_dist, n_iter=5, cv=2, n_jobs=-1, random_state=42)

# Perform randomized search on the training data (X_train_resampled, y_train_resampled)
random_search.fit(X_train_resampled, y_train_resampled)

# Print the best parameters found by Randomized Search
print("Best Parameters found: ", random_search.best_params_)

# Use the best estimator to make predictions on the test set
best_xgb_classifier = random_search.best_estimator_
y_pred_xgb_tuned = best_xgb_classifier.predict(X_test)

# Calculate accuracy for tuned XGBoost model
accuracy_xgb_tuned = accuracy_score(y_test_encoded, y_pred_xgb_tuned)
print(f'Tuned XGBoost Accuracy: {accuracy_xgb_tuned:.2f}')

# Generate and print the classification report for tuned XGBoost model
class_report_xgb_tuned = classification_report(y_test_encoded, y_pred_xgb_tuned)
print('Tuned XGBoost Classification Report:\n', class_report_xgb_tuned)
```
