```python
from imblearn.over_sampling import SMOTE
```

```python
df=pd.read_csv("spotifyfinaldatasetforML.csv")
```

```python
df
```

```python
df=df[df["year"]>2016]
```

```python
df.columns
```

```python
df=df.drop(['minor', 'key_0'], axis=1)
```

```python
df.rename(columns = {'major':'mode'}, inplace = True)
```

```python
df=df.drop_duplicates()
```

```python
df
```

```python
a=df[df["popularity"]>=70]
```

```python
a
```

```python
b = df[(df['popularity'] < 70) &(df['popularity']>=40)]
```

```python
b
```

```python
c = df[(df['popularity'] < 40) &(df['popularity']>=15)]
```

```python
c
```

```python
d=df[df["popularity"]<15]
```

```python
d
```

```python
a = a.assign(popularity='hit')
b = b.assign(popularity='popular')
c = c.assign(popularity='not_hit')
d = d.assign(popularity='flop')
```

```python
df=pd.concat([a,b,c,d])
```

```python
df.columns
```

```python
columns_to_remove = [col for col in df.columns if col=="year" or col.endswith('_ar')]

# Create a new DataFrame by excluding specified columns
new_df = df.drop(columns=columns_to_remove)
```

```python
new_df.columns
```

```python
X = new_df.drop('popularity', axis=1)  # Features
y = new_df['popularity']  # Target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, stratify=y, random_state=42)

# Apply SMOTE to the training set only
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Initialize and train the Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=50, random_state=42)
rf_classifier.fit(X_train_resampled, y_train_resampled)

# Make predictions on the test set
y_pred = rf_classifier.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Generate and print the classification report
class_report = classification_report(y_test, y_pred)
print('Classification Report:\n', class_report)
```

```python
adaboost_classifier = AdaBoostClassifier(n_estimators=50, random_state=42)
adaboost_classifier.fit(X_train_resampled, y_train_resampled)

# Make predictions on the test set
y_pred_adaboost = adaboost_classifier.predict(X_test)

# Calculate accuracy
accuracy_adaboost = accuracy_score(y_test, y_pred_adaboost)
print(f'AdaBoost Accuracy: {accuracy_adaboost:.2f}')

# Generate and print the classification report for AdaBoost
class_report_adaboost = classification_report(y_test, y_pred_adaboost)
print('AdaBoost Classification Report:\n', class_report_adaboost)
```
